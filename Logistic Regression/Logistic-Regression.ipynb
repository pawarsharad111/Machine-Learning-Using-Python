```python
def impute_age(cols):
    p=cols[0]
    a=cols[1]
    
    if pd.isnull(a):
        if p==1:
            return 38
        elif p==2:
            return 30
        elif p==3:
            return 25
        
    else:
        return a
```

```python
df['Age']=df[['Pclass','Age']].apply(impute_age, axis=1)
```
